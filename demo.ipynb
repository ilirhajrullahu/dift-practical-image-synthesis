```python
torch.cuda.set_device(0)
```

```python
dift = sd_module.SDFeaturizer()
```

```python
# you can choose visualize cat or guitar
category = random.choice(['cat', 'guitar'])

print(f"let's visualize semantic correspondence on {category}")
```

```python
if category == 'cat':
    filelist = ['/content/mydrive/MyDrive/dift/assets/cat.png', '/content/mydrive/MyDrive/dift/assets/target_cat.png', '/content/mydrive/MyDrive/dift/assets/target_cat.png']
elif category == 'guitar':
    filelist = ['/content/mydrive/MyDrive/dift/assets/guitar.png', '/content/mydrive/MyDrive/dift/assets/target_guitar.png', '/content/mydrive/MyDrive/dift/assets/target_guitar.png']

prompt = f'a photo of a {category}'
```

```python
torch.cuda.empty_cache()
```

```python
ft = []
imglist = []

# decrease these two if you don't have enough RAM or GPU memory
img_size = 768
ensemble_size = 8
torch.cuda.empty_cache()
for filename in filelist:
    img = Image.open(filename).convert('RGB')
    img = img.resize((img_size, img_size))
    imglist.append(img)
    img_tensor = (PILToTensor()(img) / 255.0 - 0.5) * 2
    ft.append(dift.forward(img_tensor=img_tensor,
                           prompt=prompt,
                           ensemble_size=ensemble_size))
ft = torch.cat(ft, dim=0)

gc.collect()
torch.cuda.empty_cache()
```

```python
demo = visualization_module.Demo(imglist, ft, img_size)
```

```python
'''
left is source image, right is target image.
you can click on the source image, and DIFT will find the corresponding
point on the right image, mark it with red point and also plot the per-pixel
cosine distance as heatmap.
'''
demo.plot_img_pairs(fig_size=5)
```
